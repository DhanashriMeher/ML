# Shuffling and creating Train and Test set

```python
from sklearn.utils import shuffle
```

```python
#shuffling dataset
data=shuffle(data,random_state=42)
data
```

```python
#creating 4 divisions
div =int(data.shape[0]/4)
div
```

```python
# 3 parts to train set and 1 part to test set
train =data.loc[:3*div+1,:]
test =data.loc[3*div+1:]
```

#Simple Mean(mean of height)

```python
#storing simple mean in a new column in the test set as "simple_mean"
test['simple_mean'] =train['Height'].mean()
```

```python
#calculating mean absolute error
from sklearn.metrics import mean_absolute_error as MAE

simple_mean_error = MAE(test['Height'],test['simple_mean'])
simple_mean_error
```

# Mean height with respect to Nationality

```python
out_type =pd.pivot_table(train,values='Height',index=['Nationality'],aggfunc=np.mean)
out_type
```

```python
#Initializing new column to zero
test['out_type_mean'] =0

#For every unique entry in outlet_Identifier
for i in train['Nationality'].unique():
#Assign the mean value corresponding to unique entry
    test['out_type_mean'][test['Nationality']==str(i)] = train['Height'][train['Nationality'] == str(i)].mean()    
```

```python
#Calculating mean absolute error
out_type_error = MAE(test['Height'],test['out_type_mean'])
out_type_error
```

Mean Item Outlet Sales with respect to joined

```python
out_year =pd.pivot_table(train,values ='Height',index= ['Joined'],aggfunc =np.mean)
out_year
```

```python
#Initializing new column to zero
test['Out_year_mean']=0

#for every unique entry in outlet identifier
for i in train['Joined'].unique():
    #Assign the mean value corresponding to unique entry
    test['Out_year_mean'][test['Joined'] == str(i)]=train['Height'][train['Joined'] == str(i)].mean()
```

```python
#Calculating mean absolute error
out_year_error = MAE(test['Height'],test['Out_year_mean'])
out_year_error
```
